### Extracting each files using user defined funtion readresume and ceating a dataframe called df

```python
def readresume(directory):
    file_path = []
    category = []
    file_content = []
    file_name = []

    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path.append(os.path.join(root, file))
            folder_name = os.path.basename(root)
            category.append(folder_name)
            file_name.append(file)

            # Extract the file content using textract
            content = textract.process(os.path.join(root, file)).decode('utf-8')
            file_content.append(content)

    # Create a DataFrame
    df = pd.DataFrame({
        'File Name': file_name,
        'File Content': file_content,
        'Category': category
    })

    return df
```

```python
import os
import textract
from docx import Document
import xml.etree.ElementTree as ET
import PyPDF2

# Specify the directory containing the files
directory = "E:\\Project 257\\Codebook Mine\\for group\\Resumes"

# Iterate over the files in the directory
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)

    # Check if the file is a text-based file (e.g., PDF, DOCX, TXT)
    if file_path.endswith(('.pdf', '.docx', '.doc', '.txt', '.xml')):
        if file_path.endswith('.pdf'):
    # Use PyPDF2 to extract text from PDF files

            with open(file_path, 'rb') as file:
                pdf_reader = PyPDF2.PdfReader(file)
                text = ''
                for page in pdf_reader.pages:
                    text += page.extract_text()

        elif file_path.endswith(('.txt')):
            # Use textract to extract text from PDF or TXT files
            text = textract.process(file_path).decode('utf-8')
        elif file_path.endswith('.docx'):
            # Use python-docx to extract text from DOCX files
            doc = Document(file_path)
            paragraphs = [paragraph.text for paragraph in doc.paragraphs]
            text = '\n'.join(paragraphs)
        elif file_path.endswith('.doc'):
        # Use textract to extract text from old DOC files
            text = textract.process(file_path, method='tesseract', encoding='utf-8')

        elif file_path.endswith('.xml'):
            # Use XML parsing to extract text from XML files
            tree = ET.parse(file_path)
            root = tree.getroot()
            text = ' '.join([element.text for element in root.iter() if element.text])
        else:
            text = "Unsupported file type"

    RESUMEDF=RESUMEDF.append({'Filename':file_name,'content':text},ignore_index=True)

        # Print or save the extracted text
    print("File:", filename)
    print("Text:")
    print(text)
    print("-------------------------------")
```

```python
import subprocess

def extract_text_from_doc(file_path):
    command = ['antiword', file_path]
    process = subprocess.Popen(command, stdout=subprocess.PIPE)
    output, error = process.communicate()
    text = output.decode('utf-8')
    return text

# Example usage:
file_path = 'Resumes/React JS Developer_AnjaniPriyadarshini.doc'
text = extract_text_from_doc(file_path)
print(text)
```
